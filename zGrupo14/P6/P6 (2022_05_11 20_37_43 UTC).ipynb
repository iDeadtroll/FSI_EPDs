```python
import pandas as pd
import matplotlib.pyplot as plt
df_futbol = pd.read_csv("14_FootballClubs.csv",encoding = "ISO-8859-1")
plt.figure(figsize = (150,100))
grafico = df_futbol.sort_values(by = ['Market_Value_Of_Club'],ascending = False)
plt.bar(grafico.Club_Name[:20],grafico.Average_Age_Of_Players[:20])
plt.xlabel("Clubes",fontsize = 100)
plt.ylabel("Valor del club en el mercado",fontsize = 100)
plt.title("Los 20 equipos con mayor valor en el mercado",fontsize = 300)
plt.savefig("grafico.png")

```

# En la grfica presente podemos observar como se reflejan los 20 clubes que poseen los mayores valores en el mercado. Para ello hemos utilizado una grfica de barras, en el eje Y vemos el valor y en el eje X vemos los Clubes #
